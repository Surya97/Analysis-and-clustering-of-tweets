```python
import pandas as pd
tweets = pd.read_csv('analysed_tweets.csv',header=0, encoding="utf8")
```

```python
tweets.head()
```

```python
#remove retweets by using the RT string which is present in all text field of the tweet
tweets=tweets[tweets.text.str.contains("RT")==False]
```

```python
tweets.head()
#see below that all those retweets which appeared above are dropped.
```

```python
# but due to dropping those dataframe entries, disturbance in index occurred.
# so resetting the index 
tweets = tweets.reset_index(drop=True)
```

```python
tweets.head()
# the index is back to normal
```

```python
#Now this is the dataset we need to work on as we don't have any retweets. So saving this as 'tweets_dataframe.csv'
'''Point to be noted is the generated index will also be saved if we don't mention index=False explicitly and when when we try to
print the dataframe after reading from csv file we get another column 'unamed' with the generated indexes.''' 
tweets.to_csv('tweets_dataframe.csv',index=False, sep='\t', encoding='utf-8')
```
