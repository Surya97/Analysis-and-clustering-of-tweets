```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('demonetization.csv',usecols=["text", "id"])
df.head()
```

```python
rows=len(df)
rows
```

```python
all_tweets=[]
for i in xrange(rows):
    all_tweets.append(df.loc[i]['text'])
```

```python
import re
```

```python
#remove @
for i in xrange(len(all_tweets)):
    new_tweet=[]
    each=all_tweets[i]
    at_the_rate = re.findall(r'(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([^a-zA-Z0-9-_\.])',each)
    urls = re.findall('https://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', each)
    for word in each.split():
        if word not in at_the_rate and word not in urls:
            new_tweet.append(word)
    new_tweet = ' '.join(new_tweet)
    all_tweets[i]=new_tweet

all_tweets
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words('english'))
len(stopwords)
```

```python
for i in xrange(len(all_tweets)):
    new_tweet=[]
    each=(all_tweets[i]).lower()
    words=re.findall(r'\w+',each)
    for word in each.split():
        if word not in stopwords:
            new_tweet.append(word)
    new_tweet = ' '.join(new_tweet)
    all_tweets[i]=new_tweet

all_tweets
```
