```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tweets_dataframe.csv', sep="\t")
```

```python
df.head()
```

```python
rows=len(df.index)
print rows
```

```python
df.shape
```

```python
import re as reg
```

```python
#remove @.*:
# regex @\w+[" ",:] " " for tweets with only @... 
for i in xrange(rows):
    tweet=str(df.loc[i]['text'])
    remove_words=reg.findall(r'(?:\S@\S*|http(?=.*://)\S*)',tweet)
    new_tweet=""
    for word in tweet.split():
        if word not in remove_words:
            new_tweet+=word+' '
    df.replace(df.loc[i]['text'], new_tweet, inplace=True)
        
```

```python
df
```

```python
# make all tweets into lower case
for i in xrange(rows):
    tweet=str(df.loc[i]['text']).lower()
    df.replace(df.loc[i]['text'], tweet, inplace=True)
    
```

```python
df.head()
```

```python
df.to_csv('tweets_dataframe.csv', index=False, sep='\t', encoding='utf-8')
```
